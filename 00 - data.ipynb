```
import pandas as pd
import os
import random
import urllib.request
import json
import time

URL_customers = "https://raw.githubusercontent.com/sergioballesterossolanas/demo_resources/master/customers.csv"
URL_orders = "https://raw.githubusercontent.com/sergioballesterossolanas/demo_resources/master/sales_oders.json"
orders_file_name = URL_orders.split("/")[-1]
save_path_customers = "/dbfs/tmp/{}/customers".format(my_suffix)
save_path_orders = "/dbfs/tmp/{}/orders".format(my_suffix)

if not os.path.exists(save_path_customers):
    os.makedirs(save_path_customers)
if not os.path.exists(save_path_orders):
    os.makedirs(save_path_orders)
```

```
urllib.request.urlretrieve(URL_orders, orders_file_name)

with open(orders_file_name) as f:
    lines = f.read().splitlines()
    
df_customers = pd.read_csv(URL_customers)
df_customers.to_csv(save_path_customers + "/{}.csv".format("mydata"), index = False)
```

```
for i in range(100000):
    print(i)
    time.sleep(20)
    try:
        with open(save_path_orders + "/{}.json".format(i), 'w+') as f:
            json_loads = json.loads(random.choice(lines))
            json.dump(json_loads, f)
    except:
        pass
```
